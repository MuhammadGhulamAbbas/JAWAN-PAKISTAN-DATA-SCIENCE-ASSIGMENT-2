```python
attack_mean = data["Attack"].mean()

def set_attack(val):
    if val < attack_mean:
        return "Attack Low"
    elif val == attack_mean:
        return "Attack neutral"
    else:
        return "Attack High"
```

```python
# "Attack_high_low"  -> "Attack High" , "Attack Low"
data["Attack_high_low"] = data["Attack"].apply(set_attack)
data
```

column of speed high and speed low


```python
Speed_mean = data["Speed"].mean()

def set_speed(val):
    if val < Speed_mean:
        return "Speed Low"
    elif val == Speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"

data["High_Low_Speed"] = data["Speed"].apply(set_speed)
data.head()
```

```python
HP_mean = data["HP"].mean()

def set_HP(val):
    if val < HP_mean:
        return "HP Low"
    elif val == HP_mean:
        return "HP neutral"
    else:
        return "HP High"

data["High_Low_HP"] = data["HP"].apply(set_HP)
data.head()
data
```
